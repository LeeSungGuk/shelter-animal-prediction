### Nan Data 삭제

```python
x_data = load_data.dropna()
len(x_data)
```

```python
data1 = []
for i in x_data['AgeuponOutcome'] :
    data = i.split(' ')
    #print(data)
    if('year' in data[1] ):
        data1.append(int(data[0]) * 365)
    elif('month' in data[1]):
        data1.append(int(data[0]) * 31)
    elif('week' in data[1]):
        data1.append(int(data[0]) * 7)
    elif('day' in data[1]):
        data1.append(int(data[0]))
        
len(data1)
```

```python
def f(data):
    if('year' in data):
        return int(data) * 365
    elif('month' in data):
        return int(data) * 31
    elif('week' in data):
        return int(data) * 7
    elif('day' in data):
        return data    
```

```python
a = lambda x : f()
x_data['AgeuponOutcome'].apply(a)
```
